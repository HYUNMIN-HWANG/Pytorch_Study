# Logistic Regression 
이름이 '회귀'지만, 실제로는 분류 작업에 사용

## 이진분류 Binary classification


```python
## Sigmoid
# 출력값을 0과 1 사이의 값으로 조정하여 반환함
# x = 0일 때, y = 0.5의 값을 갖는다.

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, 'g')
plt.plot([0,0], [1.0, 0.0], ':')    # [가로 범위],[세로 범위] 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
### W값의 변화에 따른 경사도의 변화
# W값이 커질수록 경사도가 커진다. 
# 여기서는 W 값이 직선의 기울기가 아닌, 그래프의 경사도를 결정한다.

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x, y1, 'r', linestyle='--')    # W의 값이 0.5일 때
plt.plot(x, y2, 'g')                    # W의 값이 1일 때
plt.plot(x, y3, 'b', linestyle='--')    # W의 값이 2일 때
plt.plot([0,0],[1.0,0.0],':')           # [가로 범위],[세로 범위] 가운데 점선 추가
plt.title("Sigmoid Function")
plt.show()

```

```python
## b값의 변화에 따른 좌, 우 이동

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x + 0.5)
y2 = sigmoid(x + 1)
y3 = sigmoid(x + 1.5)

plt.plot(x, y1, 'r', linestyle='--')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b', linestyle='--')
plt.plot([0,0],[1.0,0.0],':')
plt.show()
```

## 시그모디느 함수를 이용한 분류
시그모이드 함수의 출력값은 0과 1로 수렴한다. 이런 특징을 사용하여 분류 작업에 사용할 수 있다.          
만약 임계값이 0.5일 때, 출력값이 0.5이상이면 1(true), 0.5이하면 0(false) 로 판단한다.
